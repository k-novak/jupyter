```python
%matplotlib inline
import statistics
import random
from matplotlib import pyplot as plt
import numpy as np
```

```python
# paremeters for the random number generator
mean = 0
stdev = 0.5
```

```python
data = []
for i in range(1000):
    data.append(random.gauss(mean, stdev))
```

```python
# plt.rcParams['figure.figsize'] = (20,10)
fig = plt.figure()
ax=fig.add_subplot(111)
ax.set_xlim(left=-3, right=3)
ax.set_ylim(top=.1, bottom=-.1)
ax.boxplot(data, vert=False)

y = data
x = np.random.normal(0, 0.04, size=len(y))
ax.plot(y, x, 'r.', alpha=0.2)

plt.show()
```

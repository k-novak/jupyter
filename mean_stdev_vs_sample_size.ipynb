```python
%matplotlib inline
import statistics
import random
from matplotlib import pyplot as plt
import numpy as np
```

```python
# paremeters for the random number generator
mean = 0
stdev = 0.5
USL = 3
LSL = -3
```

```python
data = []
STDEV = []
MEAN = []
CPK = []

for i in range(1000):
    data.append(random.gauss(mean, stdev))

x = []
for i in range(2, 1000):
    x.append(len(data[:i]))
    m = statistics.mean(data[:i])
    s = statistics.stdev(data[:i])
    c = min((USL-m)/(3*s), (m-LSL)/(3*s))
    
    MEAN.append(m)
    STDEV.append(s)
    CPK.append(c)
    
fig = plt.figure()
ax=fig.add_subplot(111)

p1 = ax.plot(x, MEAN)
p2 = ax.plot(x, STDEV)
# p3 = ax.plot(x, CPK)

# ax.legend((p1[0], p2[0], p3[0]), ('mean', 'stdev', 'cpk'))
ax.legend((p1[0], p2[0]), ('mean', 'stdev'))
ax.grid()
```
